```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
import numpy as np

class SVMclassifier():

  def __init__(self,ni,l,lamdap):
    self.ni = ni;
    self.l = l;
    self.lamdap = lamdap;

  def fit(self, X , Y):
    self.m , self.n = X.shape
    self.X = X
    self.Y = Y

    self.w = np.zeros(self.n)
    self.b = 0

    for i in range(self.ni):
      self.update_weights()

  def update_weights(self):

    y_label = np.where(self.Y <= 0 , -1 , 1)

    for index , x_i in enumerate(self.X):
      condition = y_label[index] * (np.dot(x_i , self.w) - self.b) >= 1

      if (condition == True ):

        dw = 2*self.lamdap*self.w
        db = 0

      else:
        dw = 2*self.lamdap*self.w - np.dot(x_i , y_label[index])
        db = y_label[index]

      self.w = self.w - self.l*dw
      self.b = self.b - self.l*db


  def predict(self,X):
    output = np.dot(X,self.w) - self.b
    predicted_labels = np.sign(output)
    y_hat = np.where(predicted_labels <= -1 , 0 , 1)

    return y_hat
```

```python
X = data.drop(columns='Outcome',axis=1)
Y = data['Outcome']
```

```python
scaler = StandardScaler()
scaler.fit(X)
standardized_data = scaler.transform(X)
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(standardized_data,Y,test_size=0.2,random_state=2)
```

```python
model = SVMclassifier(1000,0.01,0.001)
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)
```

```python
print('Accuracy score on training data : ',training_data_accuracy*100,'%')
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test,X_test_prediction)
print('Accuracy score on test data : ',test_data_accuracy*100,'%')
```

```python
inputdata = data.iloc[0, :-1]
inputasarray = np.asarray(inputdata)
inputreshaped = inputasarray.reshape(1,-1)
std_data = scaler.transform(inputreshaped)
prediction = model.predict(std_data)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
