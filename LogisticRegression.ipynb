```python
import numpy as np

class Logistic_Regression:
  def __init__(self,noi,lr):
    self.noi = noi
    self.lr = lr

  def fit(self,X,Y):
    self.m , self.n = X.shape
    self.w = np.zeros(self.n) # Initialize weights with the number of features
    self.b = 0


    self.X = X
    self.Y = Y
    for i in range(self.noi):
      self.update_weights()

  def update_weights(self):

    yc = 1/(1+ np.exp(-(self.X.dot(self.w) + self.b)))

    #derivatives

    dw = (1/self.m)*np.dot(self.X.T,(yc-self.Y))
    db = (1/self.m)*np.sum(yc-self.Y)

    self.w = self.w - self.lr*dw
    self.b = self.b - self.lr*db

  def predict(self,X):
      yp = 1/(1+ np.exp(-(X.dot(self.w) + self.b)))
      yp_class = [1 if i > 0.5 else 0 for i in yp]
      return yp_class
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('/content/drive/MyDrive/Dataset/Copy of diabetes.csv')
data.head()
```

```python
# Split the data into features (X) and target (Y)
X = data.drop('Outcome', axis=1)
Y = data['Outcome']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = Logistic_Regression(noi=100, lr=0.01)

model.fit(X_train_scaled, Y_train.values)

train_predictions = model.predict(X_train_scaled)
test_predictions = model.predict(X_test_scaled)

train_accuracy = accuracy_score(Y_train, train_predictions)
test_accuracy = accuracy_score(Y_test, test_predictions)

print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)
```
